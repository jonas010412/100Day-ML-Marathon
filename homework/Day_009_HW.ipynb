# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
dtype_df = app_train.dtypes.reset_index()
dtype_df.columns = ["Count", "Column Type"]
dtype_df = dtype_df.groupby("Column Type").aggregate('count').reset_index()
dtype_df
```

```python
# 先篩選數值型的欄位


int_features = []
float_features = []
object_features = []
for dtype, feature in zip(app_train.dtypes, app_train.columns):
    if dtype == 'float64':
        float_features.append(feature)
    elif dtype == 'int64':
        int_features.append(feature)
    else:
        object_features.append(feature)
print(f'{len(int_features)} Integer Features : {int_features}\n')
print(f'{len(float_features)} Float Features : {float_features}\n')
print(f'{len(object_features)} Object Features : {object_features}')


app_train.select_dtypes(['float64','int64'])
dtype_select = ['float64','int64']

numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("Numbers of remain columns" , len(numeric_columns))

# 檢視這些欄位的數值範圍
for column in numeric_columns:
    print(column)
    plt.figure(figsize=(5,3))
    app_train[column].hist(bins=40)
    plt.title(column)
    plt.show()
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
def ecdf(data):
    xs = np.sort(data)
    ys = np.arange(1, len(xs)+1)/float(len(xs))
    return xs, ys
```

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)

xs, ys = ecdf(app_train['AMT_INCOME_TOTAL'])
plt.figure(figsize=(14,4))
plt.plot(xs, ys,label="AMT_INCOME_TOTAL")
plt.title('ECDF of AMT_INCOME_TOTAL')
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([xs.min(), xs.max()]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.legend()
plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.figure(figsize=(14,4))
plt.plot(np.log(xs), ys,label="AMT_INCOME_TOTAL")
plt.title('ECDF of AMT_INCOME_TOTAL')
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.legend()
plt.show()
```
